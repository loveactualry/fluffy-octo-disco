```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
# 특정 컬럼 가져오기
df[['sepal_width', 'sepal_length', 'species']]
columns = ['sepal_width', 'sepal_length', 'species']
df[columns]
df['sepal_width']  # df.sepal_width
```

```python
df.filter(regex='length$')  # 정규 표현식으로
df.filter(regex='^sepal')  # 특정 컬럼으로 시작.
# '^x[1-5]$'

df.filter(regex='^s')  # 시작
df.filter(regex='es$')  # 끝
df.filter(regex='^(?!species$).*')  # 특정 문자는 제외하고 가져와라.
```

```python
df.loc[2:5,'sepal_width':'petal_width'] # 특정 행에 대해 데이터를 가져오고 개수만큼 보여주기.
# 행에 대한 인덱스, 열 지정.
```

```python
df.iloc[:5,[1,3]] # 인덱스 순서대로. iloc 하는 경우는 그 전까지
df.iloc[-5:,[1,3]] # 뒤에서. loc와 차이는 iloc는 그전까지만 가져온다.
```

```python
df.loc[df['sepal_length'] > 3, ['sepal_width','petal_width']]
# , 기준 앞이 행 뒤가 열
```
